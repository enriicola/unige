```python
import pulp

my_lp_problem = pulp.LpProblem("My LP Problem", pulp.LpMinimize)

# Variables
x = pulp.LpVariable('x', lowBound=0, cat='Continuous')
y = pulp.LpVariable('y', lowBound=0, cat='Continuous')

# Objective function
my_lp_problem += - y + 1/2*x, "Z"

# Constraints
my_lp_problem += y - x <= 0
my_lp_problem += y - 1 <= 0
my_lp_problem += y + x - 4 <= 0

# Solve
my_lp_problem.solve()
print("Status:", pulp.LpStatus[my_lp_problem.status])

# Print the solution
for variable in my_lp_problem.variables():
    print("{} = {}".format(variable.name, variable.varValue))

print("Z = {}".format(pulp.value(my_lp_problem.objective)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Definizione del primo Quadrante
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
```

```python
# Definizione dei vincoli

# Vincolo v1 y -x <= 0
y1 = x
# Vincolo v2 y -1 <= 0
y2 = 1
# Vincolo v3 y + x - 4 <= 0
y3 = -x + 4 

# Definizione della funzione obiettivo
y4 = 1/2*x 
```

```python
# Plotting
plt.plot(x, y1, label=r'$y - x \leq 0$', color='g')
plt.hlines(y=1, xmin=0, xmax=100, label=r'$y -1 <= 0$', color='k')
plt.plot(x, y3, label=r'$y + x - 4 \leq 0$', color='b')
plt.plot(x, y4, label=r'$y = \frac{1}{2}x$', color='r')
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
# Calcolo della regione ammissibile
X, Y = np.meshgrid(x, y)
Z = (Y>=0) & (Y - X<= 0) & (Y - 1<=0) & (Y + X - 4<=0)
plt.imshow(Z, extent=(x.min(),x.max(),y.min(),y.max()), origin="lower", cmap="Greys", alpha = 0.3)
plt.show()
```
